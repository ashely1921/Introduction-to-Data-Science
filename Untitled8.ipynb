```
import pandas as pd
d = pd.read_excel ("456.xlsx")
d.head(5)
```

```
d_107 = d[d["year"] == 107]
d_107.shape
```

```
d_107.values
```

```
d_107
```

```
d_107_bond = d_107["Government bond"] 
d_107_bond.values
```

```
grouped = d_107.groupby("country")
name = grouped["Government bond"].count().sort_values(ascending=False)
name.index
```

```
country_name = ["Taipei", "New taipei", "Taoyuan", "Taichung", "Tainan", "Kaohsiung", "Miaoli"]
```

```
d.pivot_table(
    values=['Government bond'], 
    index='country', 
    columns=['year'])
```

```
import matplotlib.pyplot as plt
country_bond_107 = len(d_107_bond.values)
plt.bar(range(country_bond_107), d_107_bond.values)
plt.xticks(range(country_bond_107), country_name)
plt.title("How many bond in each county")
plt.xlabel("country")
plt.ylabel("government bond")
plt.show()
```

```
import matplotlib.pyplot as plt
d_107_bond = d_107["Government bond"] 
plt.hist(d_107_bond)
plt.show()
```

```
Taipei = d[d["country"] == "Taipei"]
Miaoli = d[d["country"] == "Miaoli"]
plt.plot(Taipei["year"], Taipei["Government bond"], label = "Taipei")
plt.plot(Miaoli["year"], Miaoli["Government bond"], label = "Miaoli")
plt.legend()
plt.show()
```

```
for country in country_name:
  subset_df = d[d["country"] == country]
  plt.plot(subset_df["year"], subset_df["Government bond"], label = country)
plt.legend()
plt.xlabel("Year")
plt.ylabel("Government bond")
plt.title("country's government bond over 98-107")
plt.show()
```

```
for country in country_name:
  subset_df = d[d["country"] == country]
  plt.plot(subset_df["year"], subset_df["Government tax"], label = country)
plt.legend()
plt.xlabel("Year")
plt.ylabel("Government tax")
plt.title("country's government tax over 98-107")
plt.show()
```

```
for country in country_name:
  subset_df = d[d["country"] == country]
  plt.plot(subset_df["year"], subset_df["Government spending"], label = country)
plt.legend()
plt.xlabel("Year")
plt.ylabel("Government spending")
plt.title("Government spending's government tax over 98-107")
plt.show()
```
